```python
import pandas as pd
from joblib import load
from tensorflow.keras.models import load_model
```

```python
testing_file_path = 'data_testing/data_testing_PepPI.csv'
data_testing = pd.read_csv(testing_file_path)
data_testing
```

```python
# get all columns of data_testing
data_testing.columns.tolist()
```

```python
X_test = data_testing.drop(columns=['label', 'pdb_chain', 'seq', 'Uniprot_id', 'prot_seq'])
X_test
```

```python
model_path = "best_model_auc_0.8748.h5"
loaded_model = load_model(model_path)
```

```python
# Prediksi probabilitas
predictions_probabilities = loaded_model.predict(X_test)
```

```python
threshold = 0.9
predictions_classes = (predictions_probabilities >= threshold).astype(int)

results_df = data_testing[['pdb_chain', 'seq', 'Uniprot_id', 'prot_seq']].copy()
results_df['predict_probability'] = predictions_probabilities.flatten()
results_df['predict'] = predictions_classes.flatten()
results_df
```

```python
results_df = results_df.sort_values(by='predict_probability', ascending=False).reset_index(drop=True)
results_df
```
